## 9.2

Define a generator function called get_odds() that returns the odd numbers from range(10). Use a for loop to find and print the third value returned.

```python
def get_odds():
    for number in range(10):
        if number % 2 == 1:
            yield number

count = 0
for num in get_odds():
    count += 1
    if count == 3:
        print(num)
        break
```
